```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from dataset_loader import DatasetLoader
from image_generator import ImageGenerator
```

```python

base_dir = os.getcwd()  
dataset_path = os.path.join(base_dir, "dataset")
save_path = os.path.join(base_dir, "save_folder")
```

```python
dataset_loader = DatasetLoader(dataset_path)

generator = ImageGenerator(dataset_loader, num_imgs=10, min_cells=5, max_cells=25, seed=None)
generator.generate_and_save(save_path)

def show_images_from_dir(directory, num_images=3):
    image_files = sorted([f for f in os.listdir(directory) if f.endswith('.png')])[:num_images]

    plt.figure(figsize=(12, 4))
    for idx, image_file in enumerate(image_files):
        img = Image.open(os.path.join(directory, image_file))
        plt.subplot(1, num_images, idx + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f" {idx + 1}")
    
    plt.show()

show_images_from_dir(save_path, num_images=6)
```
