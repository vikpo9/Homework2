```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

```python
df_male = df[df['Sex'] == 'male']
df_female = df[df['Sex'] == 'female']
```

```python
from IPython.display import display 
display(df_male)
display(df_female)
```

```python
male=df_male.dropna()
female=df_female.dropna()
display(df_male.dropna())
display(df_female.dropna())
```

```python
male['Age'].hist(bins = 20)
female['Age'].hist(bins = 20)
```
